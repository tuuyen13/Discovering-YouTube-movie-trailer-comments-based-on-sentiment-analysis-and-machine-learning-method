```python
# Initialize variables for pagination
next_page_token = None
video_ids = []

while True:
    # Make a request to the API to retrieve video IDs from the playlist
    results = youtube.playlistItems().list(
        part="contentDetails",
        playlistId=playlist_id,
        maxResults=50,  # Maximum is 50
        pageToken=next_page_token
    ).execute()

    # Extract video IDs and append them to the video_ids list
    for item in results['items']:
        video_ids.append(item['contentDetails']['videoId'])

    # Check if there are more pages
    next_page_token = results.get('nextPageToken')

    # If there are no more pages, break out of the loop
    if not next_page_token:
        break
```

```python
# video = 'FwJf-CzMpok' , scrawl all comments from this video
# scrape comments from a video

comment_id = []
author = []
text = []
likes = []
published_at = []
updated_at = []
video_id = []
parent_id = []

    # Initialize variables for pagination

for i in range(290,300):
    videoID = video_ids[i]

    next_page_token = None
    while True:
        # Make a request to the API
        response = youtube.commentThreads().list(
            part="snippet, replies",
            videoId=videoID,
            maxResults=50,
            pageToken=next_page_token
        ).execute()

        # Extract comments and its replies from the response object
        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']
            comment_id.append(item['id'])
            author.append(comment['authorDisplayName'])
            text.append(comment['textDisplay'])
            likes.append(comment['likeCount'])
            published_at.append(comment['publishedAt'])
            updated_at.append(comment['updatedAt'])
            video_id.append(comment['videoId'])
            # check if there is a reply, if no, append 'NA' to parent_id
            if 'parentId' not in item['snippet']:
                parent_id.append('NA')
            else: parent_id.append(item['snippet']['parentId'])
         # Extract replies
            try:
                replies = item['replies']['comments']
                for reply in replies:
                    reply = reply['snippet']
                    comment_id.append(reply['parentId'])
                    author.append(reply['authorDisplayName'])
                    text.append(reply['textDisplay'])
                    likes.append(reply['likeCount'])
                    published_at.append(reply['publishedAt'])
                    updated_at.append(reply['updatedAt'])
                    video_id.append(reply['videoId'])
                    parent_id.append(reply['parentId'])
            except:
                continue
        
        # Check if there are more pages
        next_page_token = response.get('nextPageToken')

        # If there are no more pages, break out of the loop
        if not next_page_token:
            break
```

```python
# import to dataframe
comment_rep = pd.DataFrame({'comment_id': comment_id,
                   'author': author,
                   'text': text,
                   'likes': likes,
                   'published_at': published_at,
                   'updated_at': updated_at,
                   'video_id': video_id,
                   'parent_id': parent_id
                   })
```
