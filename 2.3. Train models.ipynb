```python
import pandas as pd
df = pd.read_csv('/content/drive/Shareddrives/PHÂN TÍCH DỮ LIỆU WEB/Đồ án/TECH/Model/Toan test/DataWithLabel.csv')
```

```python
# new train test list for predicting polarity
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['Comment'], df['polarity'], test_size=0.2, random_state=42)
```

```python
#vectorize
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(min_df=5, ngram_range=(1, 2), max_features=10000)
X_train_vectorized = tfidf.fit_transform(X_train)
X_test_vectorized = tfidf.transform(X_test)
```

```python
# encode y_train and y_test
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
y_train_code = label_encoder.transform(y_train)
y_test_code = label_encoder.transform(y_test)
```

# Logistic Regression


```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier

logPol = LogisticRegression(C=1.0, max_iter=100, penalty='l2', solver='lbfgs', class_weight=None)

logPolOvr = OneVsRestClassifier(logPol, n_jobs=-1)
```

```python
# cross validation
from sklearn.model_selection import cross_validate

scoring = ['accuracy', 'f1_macro', 'recall_macro', 'precision_macro']
scores=cross_validate(logPolOvr, X_train_vectorized, y_train_code, cv=10, scoring=scoring, return_train_score=True)

# print scores
print ("Accuracy: %0.2f (+/- %0.2f)" % (scores['test_accuracy'].mean(), scores['test_accuracy'].std() * 2))
print ("Precision: %0.2f (+/- %0.2f)" % (scores['test_precision_macro'].mean(), scores['test_precision_macro'].std() * 2))
print ("Recall: %0.2f (+/- %0.2f)" % (scores['test_recall_macro'].mean(), scores['test_recall_macro'].std() * 2))
print ("F1: %0.2f (+/- %0.2f)" % (scores['test_f1_macro'].mean(), scores['test_f1_macro'].std() * 2))
```

# Naive_bayes

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.multiclass import OneVsRestClassifier

# Define the Multinomial Naive Bayes classifier with your specified parameters
nb_classifier = MultinomialNB(alpha=10, fit_prior=False, class_prior=None)


# Wrap the classifier in the OneVsRestClassifier
nb_ovr = OneVsRestClassifier(nb_classifier, n_jobs=-1)

```

```python
# cross validation
from sklearn.model_selection import cross_validate

scoring = ['accuracy', 'f1_macro', 'recall_macro', 'precision_macro']
scores=cross_validate(nb_ovr, X_train_vectorized, y_train_code, cv=10, scoring=scoring, return_train_score=True)

# print scores
print ("Accuracy: %0.2f (+/- %0.2f)" % (scores['test_accuracy'].mean(), scores['test_accuracy'].std() * 2))
print ("Precision: %0.2f (+/- %0.2f)" % (scores['test_precision_macro'].mean(), scores['test_precision_macro'].std() * 2))
print ("Recall: %0.2f (+/- %0.2f)" % (scores['test_recall_macro'].mean(), scores['test_recall_macro'].std() * 2))
print ("F1: %0.2f (+/- %0.2f)" % (scores['test_f1_macro'].mean(), scores['test_f1_macro'].std() * 2))
```

# Decision tree

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.tree import DecisionTreeClassifier
from sklearn.multiclass import OneVsRestClassifier


dtPol= DecisionTreeClassifier(criterion='gini', min_samples_leaf= 1, min_samples_split= 2, max_depth= None, max_features = None, splitter='random')

dtPolOvr = OneVsRestClassifier(dtPol, n_jobs=-1)
```

```python
# cross validation
from sklearn.model_selection import cross_validate

scoring = ['accuracy', 'f1_macro', 'recall_macro', 'precision_macro', 'roc_auc_ovr_weighted']
scores=cross_validate(dtPolOvr, X_train_vectorized, y_train_code, cv=10, scoring=scoring, return_train_score=True)

# print scores
print ("Accuracy: %0.2f (+/- %0.2f)" % (scores['test_accuracy'].mean(), scores['test_accuracy'].std() * 2))
print ("Precision: %0.2f (+/- %0.2f)" % (scores['test_precision_macro'].mean(), scores['test_precision_macro'].std() * 2))
print ("Recall: %0.2f (+/- %0.2f)" % (scores['test_recall_macro'].mean(), scores['test_recall_macro'].std() * 2))
print ("F1: %0.2f (+/- %0.2f)" % (scores['test_f1_macro'].mean(), scores['test_f1_macro'].std() * 2))
print ("ROC AUC: %0.2f (+/- %0.2f)" % (scores['test_roc_auc_ovr_weighted'].mean(), scores['test_roc_auc_ovr_weighted'].std() * 2))
```

```python
#save decision tree model
import pickle
filename = 'dtPolOvr.sav'
pickle.dump(dtPolOvr, open(filename, 'wb'))
```

# Random forest

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.multiclass import OneVsRestClassifier

rfPol=RandomForestClassifier(n_estimators=100, criterion='gini', max_depth=None, min_samples_split=2, min_samples_leaf=1,max_features='sqrt')

rfPolOvr=OneVsRestClassifier(rfPol, n_jobs=-1)
```

```python
# cross validation
from sklearn.model_selection import cross_validate

scoring = ['accuracy', 'f1_macro', 'recall_macro', 'precision_macro']
scores=cross_validate(rfPolOvr, X_train_vectorized, y_train_code, cv=10, scoring=scoring, return_train_score=True)

# print scores
print ("Accuracy: %0.2f (+/- %0.2f)" % (scores['test_accuracy'].mean(), scores['test_accuracy'].std() * 2))
print ("Precision: %0.2f (+/- %0.2f)" % (scores['test_precision_macro'].mean(), scores['test_precision_macro'].std() * 2))
print ("Recall: %0.2f (+/- %0.2f)" % (scores['test_recall_macro'].mean(), scores['test_recall_macro'].std() * 2))
print ("F1: %0.2f (+/- %0.2f)" % (scores['test_f1_macro'].mean(), scores['test_f1_macro'].std() * 2))
```
